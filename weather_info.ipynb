```python
def fix_latitude_full_decimal(lat_text):
    # Check if the value is already in float form, which means it's already been corrected
    if isinstance(lat_text, float):
        return lat_text  # Return the float value as is
    # Split the string by '.' and rejoin everything except for the first point
    parts = lat_text.split('.')
    # If there are multiple points, reconstruct the number with only the first point as decimal separator
    if len(parts) > 2:
        fixed_lat = parts[0] + '.' + ''.join(parts[1:])
    else:
        # If there's only one or no points, just return the original string
        fixed_lat = lat_text
    return float(fixed_lat)
```

```python
# Haversine formula to calculate distance between two points on the Earth
def haversine(lat1, lon1, lat2, lon2):
    # Radius of the Earth in kilometers
    R = 6371.0
    # Convert coordinates from degrees to radians
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    # Difference in coordinates
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    # Haversine formula
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a))
    distance = R * c
    return distance
```

```python
def get_summarize(row):
    latitude=row["latitude"]
    longitude=row["longitude"]
    year=row["year"]

    if year==None: year=2015

    #DRY DAYS
    days_plus40C['Distance_km'] = days_plus40C.apply(lambda row_dp: haversine(latitude, longitude, row_dp['lat'], row_dp['lon']), axis=1)
    kms_dry_days=min(days_plus40C.Distance_km)
    subset_dry_days=days_plus40C[
                        (days_plus40C['Distance_km']==kms_dry_days)&(days_plus40C['Distance_km']==kms_dry_days)&
                        (days_plus40C['year']>=year)&(days_plus40C['year']<2025)
                            ].dry_days_hxmax40
    dry_days_hxmax40_mean=np.mean(subset_dry_days.dry_days_hxmax40)


    #FREEZE THAW
    freeze_thaw['Distance_km'] = freeze_thaw.apply(lambda row_dp: haversine(latitude, longitude, row_dp['lat'], row_dp['lon']), axis=1)
    kms_freeze_thaw=min(freeze_thaw.Distance_km)
    subset_freeze_thaw=freeze_thaw[
                        (freeze_thaw['Distance_km']==kms_freeze_thaw)&(freeze_thaw['Distance_km']==kms_freeze_thaw)&
                        (freeze_thaw['year']>=year)&(freeze_thaw['year']<2025)
                            ].freeze_thaw_tx0
    freeze_thaw_mean=np.mean(subset_freeze_thaw.freeze_thaw_tx0)

    #ICE DAYS
    ice_days['Distance_km'] = ice_days.apply(lambda row_dp: haversine(latitude, longitude, row_dp['lat'], row_dp['lon']), axis=1)
    kms_ice_days=min(ice_days.Distance_km)
    subset_ice_days=ice_days[
                        (ice_days['Distance_km']==kms_ice_days)&(ice_days['Distance_km']==kms_ice_days)&
                        (ice_days['year']>=year)&(ice_days['year']<2025)
                            ].ice_days
    ice_days_mean=np.mean(subset_ice_days.ice_days)

    #MAX TEMP
    max_temp_mean['Distance_km'] = max_temp_mean.apply(lambda row_dp: haversine(latitude, longitude, row_dp['lat'], row_dp['lon']), axis=1)
    kms_max_temp_mean=min(max_temp_mean.Distance_km)
    subset_max_temp_mean=max_temp_mean[
                        (max_temp_mean['Distance_km']==kms_max_temp_mean)&(max_temp_mean['Distance_km']==kms_max_temp_mean)&
                        (max_temp_mean['year']>=year)&(max_temp_mean['year']<2025)
                            ].max_temp_mean
    max_temp_mean_mean=np.mean(subset_max_temp_mean.max_temp_mean)

    #MINIMUM TEMP MEAN
    minimum_temp_mean['Distance_km'] = minimum_temp_mean.apply(lambda row_dp: haversine(latitude, longitude, row_dp['lat'], row_dp['lon']), axis=1)
    kms_minimum_temp_mean=min(minimum_temp_mean.Distance_km)
    subset_minimum_temp_mean=minimum_temp_mean[
                        (minimum_temp_mean['Distance_km']==kms_minimum_temp_mean)&(minimum_temp_mean['Distance_km']==kms_minimum_temp_mean)&
                        (minimum_temp_mean['year']>=year)&(minimum_temp_mean['year']<2025)
                            ].minimum_temp_mean
    minimum_temp_mean_mean=np.mean(subset_minimum_temp_mean.minimum_temp_mean)

    #MEAN CMIP6
    tg_mean_6['Distance_km'] = tg_mean_6.apply(lambda row_dp: haversine(latitude, longitude, row_dp['lat'], row_dp['lon']), axis=1)
    kms_tg_mean_6=min(tg_mean_6.Distance_km)
    subset_tg_mean_6=tg_mean_6[
                        (tg_mean_6['Distance_km']==kms_tg_mean_6)&(tg_mean_6['Distance_km']==kms_tg_mean_6)&
                        (tg_mean_6['year']>=year)&(tg_mean_6['year']<2025)
                            ].tg_mean_6
    tg_mean_6_mean=np.mean(subset_tg_mean_6.tg_mean_6)

    #TOTAL PRECIPITATION
    total_precipitation['Distance_km'] = total_precipitation.apply(lambda row_dp: haversine(latitude, longitude, row_dp['lat'], row_dp['lon']), axis=1)
    kms_total_precipitation=min(total_precipitation.Distance_km)
    subset_total_precipitation=total_precipitation[
                        (total_precipitation['Distance_km']==kms_total_precipitation)&(total_precipitation['Distance_km']==kms_total_precipitation)&
                        (total_precipitation['year']>=year)&(total_precipitation['year']<2025)
                            ].total_precipitation
    total_precipitation_mean=np.mean(subset_total_precipitation.total_precipitation)

    #WET DAYS PLUS 20MM
    wet_days_plus20mm['Distance_km'] = wet_days_plus20mm.apply(lambda row_dp: haversine(latitude, longitude, row_dp['lat'], row_dp['lon']), axis=1)
    kms_wet_days_plus20mm=min(wet_days_plus20mm.Distance_km)
    subset_wet_days_plus20mm=wet_days_plus20mm[
                        (wet_days_plus20mm['Distance_km']==kms_wet_days_plus20mm)&(wet_days_plus20mm['Distance_km']==kms_wet_days_plus20mm)&
                        (wet_days_plus20mm['year']>=year)&(wet_days_plus20mm['year']<2025)
                            ].wet_days_plus20mm
    wet_days_plus20mm_mean=np.mean(subset_wet_days_plus20mm.wet_days_plus20mm)


    return dry_days_hxmax40_mean, freeze_thaw_mean, ice_days_mean, max_temp_mean_mean, minimum_temp_mean_mean, tg_mean_6_mean, total_precipitation_mean, wet_days_plus20mm_mean
```
